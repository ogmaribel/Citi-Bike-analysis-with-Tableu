```python
df["Duration_hr"] = (df['Trip Duration']/3600) 
```

```python
df.head()
```

```python
# Milage estimates are calculated using an assumed speed of 7.456 miles per hour. 
# Trips over two hours max-out at 14.9 miles
def alert(c):
  if c['Duration_hr'] <= 2.0:
    return c['Duration_hr'] * 7.456
  else:
    return 14.9 
```

```python
df['distance'] = df.apply(alert, axis=1)
```
